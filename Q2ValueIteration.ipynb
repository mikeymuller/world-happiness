```python
def iterate(top_val, roll_val, bot_val):
    new_top_val = calculateTop(top_val, roll_val)
    new_roll_val = calculateRoll(top_val, roll_val, bot_val)
    new_bot_val = calculateBot(top_val, bot_val)

    return [new_top_val, new_roll_val, new_bot_val]

def calculateTop(top_val, roll_val):
    ignore = 3+.8*(.7*top_val + .3*roll_val)
    drive = 2 + .8*(.9*top_val + .1*roll_val)
    if drive > ignore:
        return (drive, 1)
    else:
        return (ignore, 0)

def calculateRoll(top_val, roll_val, bot_val):
    ignore = 1+.8*(bot_val)
    drive = .8*(.3*top_val + .6*roll_val + .1*bot_val)
    if drive > ignore:
        return (drive, 1)
    else:
        return (ignore, 0)

def calculateBot(top_val, bot_val):
    ignore = 1+.8*(bot_val)
    drive = .8*(.6*top_val + .4*bot_val)
    if drive > ignore:
        return (drive, 1)
    else:
        return (ignore, 0)
    
def getPolicy(pol_num):
    if pol_num == 0:
        return "Doesn't Drive"
    else:
        return "Drive"

def test():
    top_val = 0
    roll_val =0
    bot_val =0
    top_pol =0
    roll_pol =0
    bot_pol =0
    for i in range(0,60):
        vals = iterate(top_val, roll_val, bot_val)
        top_val = vals[0][0]
        top_pol = vals[0][1]
        roll_val = vals[1][0]
        roll_pol = vals[1][1]
        bot_val = vals[2][0]
        bot_pol = vals[2][1]
        print("Iteration " + str(i) + ":")
        print(str(top_val) + " , " + str(roll_val) + " , " + str(bot_val))
        print("Top policy: " + getPolicy(top_pol) + ", " +
        "Roll policy: " + getPolicy(roll_pol) + ", "
        + "Bot policy: " + getPolicy(bot_pol))

test()
```
